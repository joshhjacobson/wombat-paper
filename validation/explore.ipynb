## Notes

- MOZART surface pressure is the last element in the column; GEOS Chem is reversed
- For GEOS Chem pressure, use Met_PEDGE for consistency with Mike's previous work

```python
# MOZART hourly file
ds_mozart = xr.open_dataset("../1_transport/intermediates/MOZART/output/BasisFnsUpdated/201409/BasisFnsUpdated.mz4.h0.2014-09-01-03600.nc", decode_times=False)
ds_mozart
```

```python
ds_mozart["pressure_edge"] = (ds_mozart["P0"] * ds_mozart["hyai"] + ds_mozart["PS"] * ds_mozart["hybi"]) / 100

# surface level as first element like GEOS Chem
ds_mozart = ds_mozart.reindex(
        ilev=ds_mozart.ilev[::-1]
    )

# time as first dimension like GEOS Chem
ds_mozart = ds_mozart.transpose("time", "ilev", "lev", "lat", "lon")

ds_mozart["pressure_edge"].shape
```
